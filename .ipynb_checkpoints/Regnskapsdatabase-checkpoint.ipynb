```python
rogaland_df = pd.read_csv("rogaland_df.csv")
rogaland_df = rogaland_df.drop("Unnamed: 0", axis=1)
```

```python
rogaland_df = rogaland_df.drop_duplicates()
rogaland_df = rogaland_df[rogaland_df['bkode_avd_loepende']!=0]
rogaland_df = rogaland_df.dropna(subset=['bkode_avd_loepende'])
rogaland_df['bkode_avd_loepende'] = rogaland_df['bkode_avd_loepende'] // 1000
rogaland_df['bkode_avd_loepende'] = rogaland_df['bkode_avd_loepende'].astype(int)
```

```python
rogaland_df['bkode_avd_loepende'].unique()
```

```python
nace_df = pd.read_excel("nace.xlsx")
nace_df = nace_df.set_index("bkode_avd_loepende")
```

```python
nace_df
```

```python
rogaland_df['bkode_avd_loepende'].isnull().sum()
```

```python
utslipp = pd.read_excel("Utslipp-kopi.xlsx")
```

```python
rogaland_utslipp = pd.merge(rogaland_df, utslipp, how="left", on="bkode_avd_loepende")
```

```python
rogaland_utslipp["utslipp_tonn"] = rogaland_utslipp["vs"]*rogaland_utslipp["utslipp_per_millioner_kr_vs"]/1000
```

```python
rogaland_utslipp.loc[rogaland_utslipp["firmanavn"]=="NYE VEIER AS"]
```

```python
største_utslippspunkter_næringer = rogaland_utslipp.sort_values(["prioritert_populasjonstilhoerighet","utslipp_tonn"], ascending = False).drop_duplicates(["prioritert_populasjonstilhoerighet", "orgnr_avd"]).groupby(["prioritert_populasjonstilhoerighet"]).head(n=10)

```

```python
største_utslippspunkter = rogaland_utslipp.sort_values("utslipp_tonn", ascending = False).head(n=20).drop_duplicates("orgnr_avd")

```

```python
utslippspunkter_alle = rogaland_utslipp.sort_values("utslipp_tonn", ascending = False).drop_duplicates("orgnr_avd")

```

```python
utslippspunkter_alle.bkode_avd_loepende.sort_values().unique()
```

```python
nace_matching = pd.read_excel("nace_matching.xlsx")
nace_matching.rename(columns={'Nace': 'bkode_avd_loepende'}, inplace=True)
nace_matching = nace_matching.set_index("bkode_avd_loepende")
```

```python
utslipp_per_næring = rogaland_utslipp.drop_duplicates(["prioritert_populasjonstilhoerighet", "orgnr_avd"]).groupby(["prioritert_populasjonstilhoerighet"]).sum()
utslipp_per_nace = rogaland_utslipp.drop_duplicates(["bkode_avd_loepende", "orgnr_avd"]).groupby(["bkode_avd_loepende"]).sum()
```

```python
utslipp_per_næring
```

```python
utslipp_per_nace = pd.merge(utslipp_per_nace, nace_df, how="left", on="bkode_avd_loepende")
```

```python
utslipp_per_næringshovedområde = utslipp_per_nace.groupby(["Navn næringshovedområde"]).sum()
```

```python
utslipp_per_næringshovedområde
```

```python
writer = pd.ExcelWriter('utslipp per næring.xlsx', engine='xlsxwriter')
```

```python
# Write each dataframe to a different worksheet.
utslipp_per_næring.to_excel(writer, sheet_name='utslipp per næring')
utslipp_per_næringshovedområde.to_excel(writer, sheet_name="utslipp per næringshovedområde")
utslipp_per_nace.to_excel(writer, sheet_name="utslipp per nace")
writer.close()
```
