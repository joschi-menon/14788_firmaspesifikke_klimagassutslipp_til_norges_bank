```python
nace_df = pd.read_excel(FILES_DIR / "norske utslippsfaktorer JOSHI.xlsx")
```

```python
## to-sifret float til fem-sifret int

nace_df['sn2007'] = nace_df['NACEkode'] * 1000
```

```python
## Lag kolonner med to-, tre-, fire-, og fem-sifrete nacekoder

sub = ['to', 'tre', 'fire', 'fem']

for i in range(0, 4):
    nace_df = nace_df[nace_df['sn2007']!=0]
    nace_df = nace_df.dropna(subset=['sn2007'])
    nace_df['sn2007s_' + sub[i] + 'sifrer'] = nace_df['sn2007'] // (10**(4-i-1))
    nace_df['sn2007s_' + sub[i] + 'sifrer'] = nace_df['sn2007s_' + sub[i] + 'sifrer'].astype(int)
```

```python
## Sjekk om Nace-koden er int

for i in range(0, len(nace_df)):
    print((nace_df['NACEkode'][i]).is_integer())
```

```python
## lag df med to-sifrete nace-koder og df med alle som IKKE er to-sifret

nace_to = nace_df[nace_df['NACEkode'].apply(lambda x: x.is_integer() if pd.notnull(x) else False)]
nace_flere = nace_df[~nace_df['NACEkode'].apply(lambda x: x.is_integer() if pd.notnull(x) else False)]
```

```python
## Samme som før men fire-sifret

nace_fire = nace_flere[nace_flere['NACEkode'].apply(lambda x: (x*100).is_integer() if pd.notnull(x) else False)]
nace_flere = nace_flere[~nace_flere['NACEkode'].apply(lambda x: (x*100).is_integer() if pd.notnull(x) else False)]
```

```python
## Beregn estimert utslipp basert på vs

rdb_med_sn['utslipp_estimert_vs'] = rdb_med_sn['utslippsfaktor_vs']*rdb_med_sn['vs2022']/1000
```

```python
## Beregn estimert utslipp basert på oms

rdb_med_sn['utslipp_estimert_oms'] = rdb_med_sn['utslippsfaktor_prod']*rdb_med_sn['oms2022']/1000
```

```python
## alternativ måte å estimere utslipp på. Hvis det finnes et estimat basert på oms og vs, ta gjennomsnittet av begge. Hvis det finnes kun et estimat basert på oms, ta det. Hvis det finnes kun et estimat basert på vs, ta det. 

conditions = [
    ((rdb_med_sn['utslipp_estimert_vs']!=0) | (~rdb_med_sn['utslipp_estimert_vs'].isna())) & ((rdb_med_sn['utslipp_estimert_oms']!=0) | (~rdb_med_sn['utslipp_estimert_oms'].isna())),
    ((rdb_med_sn['utslipp_estimert_vs']==0) | (rdb_med_sn['utslipp_estimert_vs'].isna())) & ((rdb_med_sn['utslipp_estimert_oms']!=0) | (~rdb_med_sn['utslipp_estimert_oms'].isna())),
    ((rdb_med_sn['utslipp_estimert_vs']!=0) | (~rdb_med_sn['utslipp_estimert_vs'].isna())) & ((rdb_med_sn['utslipp_estimert_oms']==0) | (rdb_med_sn['utslipp_estimert_oms'].isna()))
]

values = [
    (rdb_med_sn['utslipp_estimert_vs'] + rdb_med_sn['utslipp_estimert_oms'])/2,
    rdb_med_sn['utslipp_estimert_oms'],
    rdb_med_sn['utslipp_estimert_vs']
]

rdb_med_sn['utslipp_estimert'] = np.select(conditions, values)
```

```python
rdb_med_sn[rdb_med_sn['utslipp_estimert']<0]
```

```python
## Korrigerer negative utslippsestimater. Hvis utslippsestimatet er negativ, men utslipp basert på vs positiv, ta estimatet basert på basert på vs.Hvis utslippsestimatet er negativ, men utslipp basert på oms positiv, ta estimatet basert på basert på oms. Hvis alle estimater er negative, sett estimatet lik null. 

conditions = [
    (rdb_med_sn['utslipp_estimert']>0),
    (rdb_med_sn['utslipp_estimert']<0) & (rdb_med_sn['utslipp_estimert_vs']>0),
    (rdb_med_sn['utslipp_estimert']<0) & (rdb_med_sn['utslipp_estimert_oms']>0),
    (rdb_med_sn['utslipp_estimert']<0) & (rdb_med_sn['utslipp_estimert_vs']<0) & (rdb_med_sn['utslipp_estimert_oms']<0),    
]

values = [
    rdb_med_sn['utslipp_estimert'],
    rdb_med_sn['utslipp_estimert_vs'],
    rdb_med_sn['utslipp_estimert_oms'],
    0
]

rdb_med_sn['utslipp_estimert_uten_negativ'] = np.select(conditions, values)
```
